```python
import pandas as pd
```

```python
import numpy as np
def generate_data(num_rows, num_cols, seed):
    np.random.seed(seed)
    data = np.random.randn(num_rows, num_cols)
    response = np.random.randn(num_rows)
    # min-max scaling
    data = (data - data.min()) / (data.max() - data.min())
    response = (response - response.min()) / (response.max() - response.min())
    return data, response
# save to file Domain + seed . csv
def save_data(data, seed):
    df = pd.DataFrame(data)
    df.to_csv('Domain' + str(seed) + '.csv', index=False)
    return df
```

```python
seed = 5
for seed in range(5):
    data, respone = generate_data(20, 6, seed)
    display(save_data(data, seed))
```
